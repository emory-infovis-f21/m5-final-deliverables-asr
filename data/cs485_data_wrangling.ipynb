```python
import pandas as pd
import numpy as np
```

```python
# Read in data
data = pd.read_csv('/content/drive/MyDrive/cs485/uspollution_pollution_us_2000_2016.csv')
data
```

```python
# Date initially string, convert to datetime
data['Date Local'] = pd.to_datetime(data['Date Local'])
# Date convert datetime to year-month, no day
data['Date Local'] = pd.to_datetime(data['Date Local'],format='%m%Y', errors='coerce').dt.to_period('m')
```

```python
data
```

```python
data.columns
```

```python
# Drop unnecessary columns
cleaner_data = data.drop(['Unnamed: 0', 'State Code', 'County Code', 'Site Num', 'Address', 'County', 'City'], axis=1)
cleaner_data
```

```python
# Group states by date, then perform aggregations 
cleaner_sum = cleaner_data.groupby(['Date Local','State']).sum()
cleaner_sum
```

```python
cleaner_mean = cleaner_data.groupby(['Date Local','State']).mean()
cleaner_mean
```

```python
# Export to gdrive
cleaner_mean.to_csv('/content/drive/MyDrive/cs485/cleaner_data_mean.csv')
cleaner_sum.to_csv('/content/drive/MyDrive/cs485/cleaner_data_sum.csv')
```
